```python
import pandas as pd
import altair as alt
alt.data_transformers.disable_max_rows()
alt.renderers.enable('browser')
```

```python
#Reading the dataset
df = pd.read_csv("uber_dataset.csv")

print(df.shape)
print(df.columns)
print(df.head())
print(df.info())
```

```python
#Inspecting unique values and missingness

print("\nMissing values per column:")
print(df.isna().sum())

print("\nNumber of unique values per column:")
print(df.nunique())

print("\nBooking Status values:")
print(df['Booking Status'].value_counts(dropna=False))

print("\nVehicle Type values:")
print(df['Vehicle Type'].value_counts(dropna=False))

print("\nPayment Method values:")
print(df['Payment Method'].value_counts(dropna=False))
```

```python
#Convert Date and Time to datetime

df['Date'] = pd.to_datetime(df['Date'], format='%d/%m/%Y')
df['Time'] = pd.to_datetime(df['Time'], format='%H:%M:%S')
```

```python
#Extract useful time features

df['Hour'] = df['Time'].dt.hour
df['Weekday'] = df['Date'].dt.day_name()
df['Month'] = df['Date'].dt.month

print(df[['Date','Time','Hour','Weekday','Month']].head())

completed = df[df['Booking Status'] == 'Completed']

print(completed.shape)
```

```python
#Altair Chart Bookings by Vehicle Type
vehicle_chart = alt.Chart(df).mark_bar().encode(
    x=alt.X('Vehicle Type:N', sort='-y'),
    y='count()',
    color='Vehicle Type:N',
    tooltip=['Vehicle Type','count()']
).properties(
    title='Number of Bookings by Vehicle Type',
    width=500
)

vehicle_chart
```

```python
#Bookings by Hour
hour_chart = alt.Chart(df).mark_bar().encode(
    x=alt.X('Hour:O', title='Hour of Day'),
    y='count()',
    tooltip=['Hour','count()']
).properties(
    title='Ride Demand by Hour'
)

hour_chart
```

```python
#Distance vs Booking Value
scatter_chart = alt.Chart(completed).mark_circle(size=60).encode(
    x='Ride Distance:Q',
    y='Booking Value:Q',
    color='Vehicle Type:N',
    tooltip=['Ride Distance','Booking Value','Vehicle Type']
).properties(
    title='Ride Distance vs Booking Value'
)

scatter_chart
```
